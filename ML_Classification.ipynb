```python
import pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt, nltk
pd.set_option('max_colwidth', 100, 'display.max_rows', 10)
_ = nltk.download(info_or_id=['names'], quiet=True)

LsM = nltk.corpus.names.words(fileids='male.txt')
LsF = nltk.corpus.names.words(fileids='female.txt')

# The first 20 names from each list
print(f'{len(LsM)} male names:  ', LsM[:20])
print(f'{len(LsF)} female names:', LsF[:20])
```

```python
# You're building a Pandas DataFrame from the NLTK names corpus, 
# labeling female names as 1 and male names as 0, adding a length feature, and then transposing the DataFrame. 
# Name=LsF + LsM -> Concatenation of both the names
# Label: 1 for female, 0 for male
print(f'Length of Male {len(LsM)}')
print(f'Length of Female {len(LsF)}')
df = pd.DataFrame(dict(Name=LsF+LsM, Y=[1]*len(LsF) + [0]*len(LsM)))
# print(df)
# Add a length feature
df['Len'] = df.Name.apply(len)
# print(df)
# Set the index to the name
df = df.set_index('Name')
# Transpose the DataFrame
df.T
df.sort_values('Len')[['Y']].T

# It performs three operations in sequence:
#Sort the DataFrame by the column 'Len'  
#→ shortest names first, longest names last

#Select only the 'Y' column  
#→ keeps just the gender label (1 = female, 0 = male)

#Transpose the result  
#→ rows become columns, columns become rows
print(df.drop('Y', axis=1))
# ✔️ Removes the column named 'Y'
# Returns a new DataFrame (does NOT modify df unless you use inplace=True)
# axis=1 tells Pandas you’re dropping a column, not a row
```

```python
from sklearn.model_selection import train_test_split
# Use the train_test_split() method to allocate 80% of observations for training and 20% for validation.
# This method, with both X and Y arguments specified, will return four objects: 
# training input tX, training output tY, validation input vX, and validation output vY. 
#These will have the same number of features, if splitting was done correctly.
tX, vX, tY, vY = train_test_split(df.drop('Y', axis=1), df.Y, test_size=0.2, random_state=0)
print(f'Training Input {tX}')
print(f'Training Output {tY}')
print(f'Validation Input {vX}')
print(f'Validation Output {vY}')
print('tX:', tX.shape, ',\t tY:', tY.shape)
print('vX:', vX.shape, ',\t vY:', vY.shape)
```

```python
# Upper and lower vowel letters are 'aeiouAEIOU'. You can reset the index in a dataframe, so that it's easier to use (as a column), then call the str attribute to access the string value. 
# From there you can use string's count() method to count matches of vowel letters. There are many other approaches.

df['nVowels'] = df.reset_index().Name.str.count(r'[aeiouAEIOU]').values
df.T
```

```python
# Upper and lower vowel letters are 'aeiouAEIOU'. You can reset the index in a dataframe, so that it's easier to use (as a column), 
# then apply a lambda function to check if the first character of a value is in the vowel list of letters or not.
df['nVowelFirst'] = df.reset_index().Name.apply(lambda s: s[0] in 'aeiouAEIOU').values*1
df.T
```
